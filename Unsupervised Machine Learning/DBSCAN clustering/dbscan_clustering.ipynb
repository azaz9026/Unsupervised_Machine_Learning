# DBSCAN clustering

***Density-Based Spatial Clustering of Applications with Noise. Finds core samples of high density and expands clusters from them.***
   ***Good for data which contains clusters of similar density***

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_moons

x , y = make_moons(n_samples=250, noise=0.05)

x , y
```

```python
df = {"data1": x[:,0] , "data2": x[:,1]}
```

```python
data = pd.DataFrame(df)
data
```

```python
sns.scatterplot(x='data1' , y='data2' , data = data)
plt.show()
```

```python
from sklearn.cluster import DBSCAN

db = DBSCAN(eps=0.2, min_samples=5)
data['predict'] = db.fit_predict(data)
```

```python
data.head(5)
```

```python
data.predict.unique() , data.predict.value_counts()
```

```python
sns.scatterplot(x='data1' , y='data2' , hue='predict' , data = data)
plt.show()
```
